```python
# Importing dependencies
import pandas as pd
import requests
import json
from sqlalchemy import create_engine
```

```python
# Find Chess.com usernames
url = "https://api.chess.com/pub/titled/GM"
response = requests.get(url).json()
```

```python
# "Check" response... haha
print(response)
```

```python
# Small attempt for API Query
usernames = response["players"]
player_url = "https://api.chess.com/pub/player/"

responses = []
try:
    for username in usernames:
        chess_data = requests.get(player_url + username).json()
        responses.append(chess_data)
        print(f'Chess user {username} and has {chess_data["followers"]} followers')
except:
    print("next")
```

```python
# More complicated, this time trying to get specific stat
# Example; Addind data to dataframe
usernames = response["players"]

# player_url = "https://api.chess.com/pub/player/"

# Create chess list
chess_data = []

responses = []
for username in usernames:
    player_url = "https://api.chess.com/pub/player/" + username + "/stats"
    chess_json = requests.get(player_url).json()
#     responses.append(chess_data)
    try:
        blitz_win = chess_json["chess_blitz"]["record"]["win"]
        blitz_loss = chess_json["chess_blitz"]["record"]["loss"]
        blitz_draw = chess_json["chess_blitz"]["record"]["draw"]
        bullet_win = chess_json["chess_bullet"]["record"]["win"]
        bullet_loss = chess_json["chess_bullet"]["record"]["loss"]
        bullet_draw = chess_json["chess_bullet"]["record"]["draw"]
        rapid_win = chess_json["chess_rapid"]["record"]["win"]
        rapid_loss = chess_json["chess_rapid"]["record"]["loss"]
        rapid_draw = chess_json["chess_rapid"]["record"]["draw"]
        blitz_best_rating = chess_json["chess_blitz"]["best"]["rating"]
        bullet_best_rating = chess_json["chess_bullet"]["best"]["rating"]
        rapid_best_rating = chess_json["chess_rapid"]["best"]["rating"]
        chess_data.append({"Username": username,
                          "Blitz Wins": blitz_win,
                            "Blitz Losses": blitz_loss,
                            "Blitz Draws": blitz_draw,
                            "Bullet Wins": bullet_win,
                            "Bullet Losses": bullet_loss,
                            "Bullet Draw": bullet_draw,
                            "Rapid Wins": rapid_win,
                            "Rapid Losses": rapid_loss,
                            "Rapid Draws": rapid_draw,
                            "Blitz Best Rating": blitz_best_rating,
                            "Bullet Best Rating": bullet_best_rating,
                           "Rapid Best Rating": rapid_best_rating,
                            })
        print("Data stored")
#         print(f'Chess user {username} and has {chess_data["chess_blitz"]["record"]["win"]} chess bitz wins')
    #         print(player_url)
    except:
        print("next")
        pass
```

```python
# Create Dataframe
chess_df = pd.DataFrame(chess_data)

# Display head of dataframe
chess_df.head()
```
